#### Loading shape file
Aside `.shp`, all other data format such as `.dbf` , `.prj`, `.xml`, `.shx` must be in the same directory
- **.shp**: Stores the shapefile's geometric data, like points, lines, and polygons.
- **.shx**: The shape index file; it helps in efficiently accessing the records in the .shp file.
- **.dbf**: A database file containing attribute data for each shape.
- **.prj**: Contains projection information, defining the coordinate system for the shapefile.
- **.xml**: Usually holds metadata about the shapefile, like its creation and data specifics.

These files complement each other, ensuring that the shapefile is accurately read and interpreted in GIS software. Missing any of these files can lead to errors or incomplete data representation.


## Loading GeoDataFrame

```python
# Load your GeoDataFrame

gd = gpd.read_file('nic_autoc2024023n_pl_a/nic_autoc2024023n_pl_a.shp')
```

```python
# Simple plot of all features
gd.plot()
```

```python
# Create a plot using one of the numeric columns (e.g., 'SHAPE_Area')
fig, ax = plt.subplots(1,1,  figsize=(10,5) )
cb=gd.plot(column='SHAPE_Area',ax=ax, cmap='viridis', legend=True)
plt.grid(alpha=0.1) # Display grid
plt.show() # Show the plot
```

```python
# Create the plot with a color column
fig, ax = plt.subplots(1, 1, figsize=(10,5))
plot = gd.plot(column='SHAPE_Area', cmap='viridis', ax=ax)

# Adjust colorbar to be horizontal
plt.colorbar(plot.collections[0], ax=ax, orientation='horizontal')

# Show the plot
plt.show()
```
